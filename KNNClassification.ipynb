```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model,preprocessing
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import io
data = pd.read_csv(io.BytesIO(uploaded['car.data']))
print(data.head())
```

```python
l = preprocessing.LabelEncoder()
```

```python
buying = l.fit_transform(list(data['buying']))
maint = l.fit_transform(list(data['maint']))
door = l.fit_transform(list(data['door']))
persons = l.fit_transform(list(data['persons']))
lug_boot = l.fit_transform(list(data['lug_boot']))
safety = l.fit_transform(list(data['safety']))
cls = l.fit_transform(list(data['class']))
```

```python
X = list(zip(buying,maint,door,persons,lug_boot,safety))
Y = list(cls)
```

```python
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(X,Y,test_size=0.1)
```

```python
print(x_train)
print(y_test)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(x_train,y_train)
```

```python
acc = model.score(x_test,y_test)
print(acc)
```

```python
predicted = model.predict(x_test)
```

```python
names = ["unacc","acc","good","vgood"]
for x in range(len(predicted)):
  print("Predicted Values : ", names[predicted[x]], "Input Data Set:", x_test[x],"Actual Values :", names[y_test[x]])
  n = model.kneighbors([x_test[x]],5,True)
  print(n)
```
